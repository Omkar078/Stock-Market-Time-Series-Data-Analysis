```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Importing individual Stocks CSV files

```python
path = 'C:/Users/OMKAR/Documents/PortFolio/Stock-Market-Time-Series-Data-Analysis/individual_stocks_5yr'
stock_list = ['AAPL_data.csv', 'GOOG_data.csv', 'MSFT_data.csv', 'AMZN_data.csv']
all_data = pd.DataFrame()
for file in stock_list:
    current_df = pd.read_csv(path+'/'+file)
    all_data = pd.concat([all_data,current_df])
```

### Analyzing Closing Price of all the stocks

```python
all_data.dtypes
```

#### Changing Date column datatype from object to date

```python
all_data['date'] = pd.to_datetime(all_data['date'])
```

```python
all_data.dtypes
```

#### List  all Stocks

```python
tech_list = all_data['Name'].unique()
tech_list
```

#### Visualization

```python
plt.figure(figsize = (20,12))
for i, company in enumerate(tech_list,1):
    plt.subplot(2, 2, i)
    df = all_data[all_data['Name']==company]
    plt.plot(df['date'],df['close'])
    plt.title(company)
```
